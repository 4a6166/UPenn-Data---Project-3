```python
import pandas as pd
```

```python
imported_df = pd.read_excel("Resources/2020-2021 Per Pupil Expenditures .xlsx", sheet_name=1)
imported_df
```

```python
public_schools_df = imported_df.loc[imported_df["Ctgy"]==1]
public_schools_df
```

```python
public_schools_df["Total Per Pupil Expenditure 2020-21"]= \
    public_schools_df["2020-2021 Local Personnel Per Pupil Expenditure"] + \
    public_schools_df["2020-2021 Local NonPersonnel Per Pupil Expenditure"] + \
    public_schools_df["2020-2021 State Personnel Per Pupil Expenditure"] + \
    public_schools_df["2020-2021 State NonPersonnel Per Pupil Expenditure"] + \
    public_schools_df["2020-2021 Federal Personnel Per Pupil Expenditure"] + \
    public_schools_df["2020-2021 Federal NonPersonnel Per Pupil Expenditure"]
public_schools_df
```

```python
public_schools_df_clean = public_schools_df[["AUN","LEA Name","Total Per Pupil Expenditure 2020-21"]]\
                                            .sort_values(by="AUN")
public_schools_df_clean
```

```python
import geojson

# Load boundaries file
with open("Resources/Pennsylvania_School_Districts_Boundaries.geojson") as f:
    gj = geojson.load(f)

features = gj['features']
```

```python
len(features)
```

```python
gj['features'][0]['properties']['aun_schdis']
```

```python
# Create nested for loops to match AUNs in dataframe and geojson files
# When a match is found, add the per pupil expenditure from the dataframe to the geojson
i=0
for i in range(len(features)-1):
    aun = features[i]['properties']['aun_schdis']
    
    j=0
    for j in range(len(public_schools_df_clean)-1):
        if str(public_schools_df_clean['AUN'][j]) == aun:
            features[i]['properties']['pupil_expend'] = public_schools_df_clean['Total Per Pupil Expenditure 2020-21'][j]
```

```python
# Check that the property was added
features[0]['properties']['pupil_expend']
```

```python
# Write JSON file with new data
from geojson import dump
with open('Resources/andrew.geojson', 'w') as f:
  dump(gj, f)   
```
