```python
import pandas as pd
```

```python
# Read in the data
# Note: despite the df name, this includes charter schools, but they will not be included as we use AUN to match public schools
# The df name is kept as public_schools_df to avoid rewriting all the existing code below
# (after finding out this sheet does not have 'Ctgy' column)
public_schools_df = pd.read_excel("Resources/2018-2019 Per Pupil Expenditures.xlsx", sheet_name=1)
public_schools_df
```

```python
# This code previously selected only the public school districts based on a "Ctgy" column
# But that is not available in the 2018-19 data, so commenting out this code
# public_schools_df = imported_df.loc[imported_df["Ctgy"]==1]
# public_schools_df
```

```python
# Sum the per pupil expenditures
public_schools_df["Total Per Pupil Expenditure 2018-19"]= \
    public_schools_df["2018-2019 Local Per Pupil Expenditure"] + \
    public_schools_df["2018-2019 Local NonPer Pupil Expenditure"] + \
    public_schools_df["2018-2019 State Per Pupil Expenditure"] + \
    public_schools_df["2018-2019 State NonPer Pupil Expenditure"] + \
    public_schools_df["2018-2019 Federal Per Pupil Expenditure"] + \
    public_schools_df["2018-2019 Federal NonPer Pupil Expenditure"]
public_schools_df
```

```python
# Deer Lakes SD expenditure data seems to be unreliable (drop it in the next step)
public_schools_df.loc[public_schools_df["LEA Name"] == "Deer Lakes SD"]
```

```python
# Select only the necessary columns
public_schools_df_clean = public_schools_df[["AUN","LEA Name","Total Per Pupil Expenditure 2018-19"]]

# Drop Deer Lakes SD which has unreliable data
public_schools_df_clean = public_schools_df_clean.drop(36)

# Reset index after dropping row
public_schools_df_clean.reset_index(inplace=True)

public_schools_df_clean
```

```python
import geojson

# Load boundaries file
with open("Resources/Pennsylvania_School_Districts_Boundaries.geojson") as f:
    gj = geojson.load(f)

features = gj['features']
```

```python
len(features)
```

```python
gj['features'][0]['properties']['aun_schdis']
```

```python
# Create nested for loops to match AUNs in dataframe and geojson files
# When a match is found, add the per pupil expenditure from the dataframe to the geojson
i=0
for i in range(len(features)-1):
    aun = features[i]['properties']['aun_schdis']
    
    j=0
    for j in range(len(public_schools_df_clean)-1):
        if str(public_schools_df_clean['AUN'][j]) == aun:
            features[i]['properties']['pupil_expend'] = public_schools_df_clean['Total Per Pupil Expenditure 2018-19'][j]
```

```python
# Check that the property was added
features[0]['properties']['pupil_expend']
```

```python
# Write JSON file with new data
from geojson import dump
with open('Resources/andrew.geojson', 'w') as f:
  dump(gj, f)   
```
