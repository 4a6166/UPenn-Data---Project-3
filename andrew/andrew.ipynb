```python
import pandas as pd
```

```python
# Read in the data
# Note: despite the df name, this includes charter schools, but they will not be included as we use AUN to match public schools
# The df name is kept as public_schools_df to avoid rewriting all the existing code below
# (after finding out this sheet does not have 'Ctgy' column)
public_schools_df = pd.read_excel("Resources/Full 2018-2019 Per Pupil Expenditures.xlsx", sheet_name=1)
public_schools_df
```

```python
# This code previously selected only the public school districts based on a "Ctgy" column
# But that is not available in the 2018-19 data, so commenting out this code
# public_schools_df = imported_df.loc[imported_df["Ctgy"]==1]
# public_schools_df
```

```python
# Sum the per pupil expenditures
public_schools_df["Total Per Pupil Expenditure 2018-19"]= \
    public_schools_df["2018-2019 Local Per Pupil Expenditure"] + \
    public_schools_df["2018-2019 Local NonPer Pupil Expenditure"] + \
    public_schools_df["2018-2019 State Per Pupil Expenditure"] + \
    public_schools_df["2018-2019 State NonPer Pupil Expenditure"] + \
    public_schools_df["2018-2019 Federal Per Pupil Expenditure"] + \
    public_schools_df["2018-2019 Federal NonPer Pupil Expenditure"]
public_schools_df
```

```python
# Deer Lakes SD expenditure data seems to be unreliable (drop it in the next step)
public_schools_df.loc[public_schools_df["LEA Name"] == "Deer Lakes SD"]
```

```python
# Select only the necessary columns
public_schools_df_clean = public_schools_df[["AUN","LEA Name","Total Per Pupil Expenditure 2018-19"]]

# Drop Deer Lakes SD which has unreliable data
public_schools_df_clean = public_schools_df_clean.drop(36)

# Reset index after dropping row
public_schools_df_clean.reset_index(inplace=True)

public_schools_df_clean = public_schools_df_clean[["AUN","LEA Name","Total Per Pupil Expenditure 2018-19"]]

public_schools_df_clean
```

```python
# Read in assessment data
keystone_df = pd.read_excel('Resources/2019 Keystone Exams School Level Data.xlsx', skiprows=4)
keystone_df
```

```python
# Combine advanced and proficient
keystone_df["Proficiency Percent"]=keystone_df["Percent Advanced"]+keystone_df["Percent Proficient"]

# Drop unnecessary columns
keystone_df_clean = keystone_df[["AUN","District Name", "Subject","Group", "Number Scored", "Proficiency Percent"]]

# Look only at the all students group
keystone_df_clean = keystone_df_clean.loc[keystone_df_clean['Group']=="All Students"]

# Calculate number of proficienct students
keystone_df_clean["Proficient Students"] = keystone_df["Number Scored"]*(keystone_df["Percent Proficient"]/100)
keystone_df_clean = keystone_df_clean.round({'Proficient Students':0})

keystone_df_clean
```

```python
# Get just the algebra scores
algebra_df = keystone_df_clean.loc[keystone_df_clean["Subject"]=="Algebra I"]
algebra_df = algebra_df.rename(columns={"Proficiency Percent": "Algebra Proficiency Percent",
                                        "Proficient Students": "Algebra Proficient Students"})
algebra_df
```

```python
# Some districts have multiple schools, we need to sum the student numbers
algebra_sums_df = algebra_df.groupby('AUN').sum('Algebra Proficient Students')
algebra_sums_df = algebra_sums_df[["Number Scored", "Algebra Proficient Students"]]
algebra_sums_df
```

```python
# Get just the literature scores
literature_df = keystone_df_clean.loc[keystone_df_clean["Subject"]=="Literature"]
literature_df = literature_df.rename(columns={"Proficiency Percent": "Literature Proficiency Percent",
                                              "Proficient Students": "Literature Proficient Students"})
literature_df
```

```python
# Some districts have multiple schools, we need to sum the student numbers
literature_sums_df = literature_df.groupby('AUN').sum('Literature Proficient Students')
literature_sums_df = literature_sums_df[["Number Scored", "Literature Proficient Students"]]
literature_sums_df
```

```python
# Get just the biology scores
biology_df = keystone_df_clean.loc[keystone_df_clean["Subject"]=="Biology"]
biology_df = biology_df.rename(columns={"Proficiency Percent": "Biology Proficiency Percent",
                                       "Proficient Students": "Biology Proficient Students"})
biology_df
```

```python
# Some districts have multiple schools, we need to sum the student numbers
biology_sums_df = biology_df.groupby('AUN').sum('Biology Proficient Students')
biology_sums_df = biology_sums_df[["Number Scored", "Biology Proficient Students"]]
biology_sums_df
```

```python
# Merge all subject dfs together
merged_df = algebra_sums_df.merge(literature_sums_df, on='AUN', validate='one_to_one')
merged_df = merged_df.merge(biology_sums_df, on='AUN', validate='one_to_one')
merged_df = merged_df.rename(columns={"Number Scored_x": "Algebra Students",
                                     "Number Scored_y": "Literature Students",
                                     "Number Scored": "Biology Students"})
merged_df
```

```python
# Calculate proficiency percents by school district unique identifier ('AUN')
merged_df["Percent Algebra Proficient"] = merged_df["Algebra Proficient Students"]/merged_df["Algebra Students"]*100
merged_df["Percent Literature Proficient"] = merged_df["Literature Proficient Students"]/merged_df["Literature Students"]*100
merged_df["Percent Biology Proficient"] = merged_df["Biology Proficient Students"]/merged_df["Biology Students"]*100
merged_df_clean = merged_df[["Percent Algebra Proficient", "Percent Literature Proficient", "Percent Biology Proficient"]]
rates_df = merged_df_clean.reset_index()
rates_df
```

```python
final_merged_df = public_schools_df_clean.merge(rates_df, on='AUN', validate='one_to_one')
final_merged_df = final_merged_df.round(2)
final_merged_df
```

```python
import geojson

# Load boundaries file
with open("Resources/Pennsylvania_School_Districts_Boundaries.geojson") as f:
    gj = geojson.load(f)

features = gj['features']
```

```python
len(features)
```

```python
gj['features'][0]['properties']['aun_schdis']
```

```python
# Create nested for loops to match AUNs in dataframe and geojson files
# When a match is found, add the per pupil expenditure from the dataframe to the geojson
i=0
for i in range(len(features)-1):
    aun = features[i]['properties']['aun_schdis']
    
    j=0
    for j in range(len(final_merged_df)-1):
        if str(final_merged_df['AUN'][j]) == aun:
            features[i]['properties']['pupil_expend'] = final_merged_df['Total Per Pupil Expenditure 2018-19'][j]
            features[i]['properties']['algebra'] = final_merged_df['Percent Algebra Proficient'][j]
            features[i]['properties']['literature'] = final_merged_df['Percent Literature Proficient'][j]
            features[i]['properties']['biology'] = final_merged_df['Percent Biology Proficient'][j]
```

```python
# Check that the property was added
print(features[0]['properties']['pupil_expend'])
print(features[0]['properties']['algebra'])
print(features[0]['properties']['literature'])
print(features[0]['properties']['biology'])
```
